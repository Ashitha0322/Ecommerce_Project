### TASK 3.2: Handle Missing Values

```python
Ecom_df.isnull().sum()
```

```python
# checking the misssing values percentage in each column
(Ecom_df.isnull().sum()/Ecom_df.shape[0]) * 100
```

```python
for col in Ecom_df.columns:
    missing_per=(Ecom_df[col].isnull().sum()/Ecom_df.shape[0]) * 100
    if missing_per <5:
        Ecom_df.dropna(subset=[col])
    else:
        print(f" the missing percentage is {missing_per}, so we can not drop the column.")
```

```python
#dropping missing values permanently
Ecom_df.dropna(subset=['customer_id','payment_method','city'],inplace=True)
```

## PART 4: FEATURE ENGINEERING

### TASK 4.1: Extract Date Components

```python
#converting order_id into datetime 
Ecom_df['order_date']=pd.to_datetime(Ecom_df['order_date'])
Ecom_df['delivery_date']=pd.to_datetime(Ecom_df['delivery_date'])
```

### TASK 7.2: Find Top and Bottom Records

```python
#Use nlargest() and nsmallest()
#DataFrame.nlargest(n, columns, keep='first')
```

+ DataFrame.nlargest(n, columns, keep='first')
+ n: int, Number of values to select
+ columns: Column to check for values or user can select column while calling too.
+ example: data["age"].nsmallest(3) OR data.nsmallest(3, "age")]
+ keep: object to set which value to select if duplicates exit. Options are 'first' or 'last'

```python
#Which are the top-selling and least-selling products?
Ecom_df.head(2)
```

```python
Ecom_df.groupby('product_name')['total_amount'].sum()
# here product_name is index
```

```python
Ecom_df.groupby('product_name')['total_amount'].sum().round(2).reset_index()
# using reset_index product_name is included under column not index
```

```python
#top most sold
Ecom_df.groupby('product_name')['total_amount'].sum().reset_index().nlargest(5,'total_amount')
```

```python
#least sold
Ecom_df.groupby('product_name')['total_amount'].sum().reset_index().nsmallest(5,'total_amount')
```
